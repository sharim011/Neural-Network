```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras  import layers 
from  tensorflow.keras.utils import  to_categorical
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train ,y_train) ,(x_test ,y_test) = keras.datasets.mnist.load_data()

fig,axes = plt.subplots(5,10 , figsize = (12,7) , subplot_kw = {'xticks':[],
                                                                'yticks':[]})

for i , ax  in enumerate(axes.flat):
    ax.imshow(x_train[i],cmap=plt.cm.gray_r)
    ax.text(0.45,0.05,str(y_train[i]),transform=ax.transAxes)

plt.show()
```

```python
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0
x_train = x_train.reshape(-1 , 28 , 28,1)
x_test = x_test.reshape(-1 , 28 , 28,1)
y_train = to_categorical(y_train,num_classes = 10)
y_test = to_categorical(y_test,num_classes = 10)
```

```python
#3 define the cnn model

model = keras.Sequential([layers.Conv2D(32,(3,3) , activation= 'relu' , input_shape=(28,28,1)) ,
                         layers.MaxPooling2D((2,2)),
                         layers.Conv2D(64,(3,3) , activation = 'relu'),
                         layers.MaxPooling2D((2,2)),
                         layers.Flatten(),
                         layers.Dense(10,activation='softmax')])


model.summary()
```

```python
#4  . train th model (without displying epochs)

model.compile(optimizer='adam' , loss='categorical_crossentropy' , metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10,batch_size=32,verbose=0) #set verbose=0
```

```python
# 5 evaluate  the model
loss , accuracy = model.evaluate(x_test ,y_test , verbose=0)
print('Test accuracy:' , accuracy)
```

```python
# 6 disply  a test image and make a prediction

test_image = x_test[50]
plt.tick_params(axis='both',which='both',bottom=False,top=False ,left=False,
                right=False , labelbottom=False,labelleft=False)

plt.imshow(test_image.reshape(28,28),cmap=plt.cm.gray_r)
plt.show()
```

```python
predicted_class = np.argmax(model.predict(test_image.reshape(1,28,28,1)),axis=-1)[0]
print('Look like a ' +str(predicted_class)+'!')
```

```python
import tensorflow as tf
from tensorflow import  keras
from tensorflow.keras import layers , datasets, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
(train_images,train_labels),(test_images,test_labels)= datasets.cifar10.load_data()
```

# normalize pixel  between 0 to 1 

```python
train_images,test_images =  train_images/255.0 , test_images/255.0
```

```python
class_names = ['aeroplane','automobile','bird','cat',
       
               'deer','dog','frog','horse','ship','truck']
```

```python
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i][0]])
    
plt.show()
```

```python
model = models.Sequential([layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),
                          layers.MaxPooling2D((2,2)),
                          layers.Conv2D(62,(3,3),activation='relu'),
                          layers.MaxPooling2D((2,2)),
                          layers.Conv2D(64,(3,3),activation='relu'),
                          layers.Flatten(),
                          layers.Dense(64,activation='relu'),
                          layers.Dense(10)])

model.summary()
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
history = model.fit(train_images,train_labels,epochs=10,validation_data=(test_images,test_labels),verbose=0)
```

```python
def predict_single_image(model,image_index):
    img =  test_images[image_index]
    true_label = test_labels[image_index][0]
    
    img_expanded = np.expand_dims(img,axis=0)
    
    predictions = model.predict(img_expanded,verbose=0)
    predicted_label =  np.argmax(predictions[0])
    confidence = tf.nn.softmax(predictions[0])
    
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.title(f"predicted:{class_names[predicted_label]} (confidence:{confidence[predicted_label]:2f})\n" f"true:{class_names[true_label]}")
    
    plt.show()
    
    print(f"predicted_label : {class_names[predicted_label]}")
    print(f"true_label : {class_names[true_label]}")
    print("Confidence Scores")
    
    for  i,score in enumerate(confidence):
        print(f"{class_names[i]} : {score:.4f}")

    
    
```

```python
image_index = 1
predict_single_image(model,image_index)
```

```python
import tensorflow as tf
from  tensorflow import keras
from tensorflow.keras  import Sequential , datasets
from tensorflow.keras.layers  import Conv2D , MaxPooling2D
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
(x_train , y_train) , (x_test , y_test) = datasets.mnist.load_data()
fig,axes = plt.subplots(5,10 , figsize=(10,7) , subplot_kw={'xticks':[] , 'yticks':[]})
for i ,ax in enumerate(axes.flat):
    ax.imshow(x_train[i],cmap=plt.cm.gray_r)
    ax.text(0.5,0,str(y_train[i]),transform=ax.transAxes)
    
plt.show()
```

```python
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0
x_train = x_train.reshape(-1,28,28,1)
x_test = x_test.reshape(-1,28,28,1)
y_train = to_categorical(y_train ,num_classes=10)
y_test = to_categorical(y_test ,num_classes=10)
```

```python
from tensorflow.keras import layers


model = Sequential([Conv2D(32,(3,3) , activation = 'relu' , input_shape=(28,28,1)),
                   MaxPooling2D((2,2)),
                   Conv2D(64,(3,3) ,activation='relu'),
                   MaxPooling2D((2,2)),
                   layers.Flatten(),
                   layers.Dense(10,activation='softmax')])

model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10,batch_size=32,verbose=0)
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
print(f"accuracy_score : {accuracy*100}")
```

```python
test_image = x_test[4]
plt.tick_params(axis='both',which='both',bottom=False,top=False,left=False,right=False,
               labelbottom=False,labelleft=False)

plt.imshow(test_image.reshape(28,28,1),cmap=plt.cm.gray_r)
plt.show()
```

```python
predicted_no = np.argmax(model.predict(test_image.reshape(-1,28,28,1)),axis=-1)[0]
print(f"predicted_no : ",{predicted_no})
```

```python
(train_images,train_labels),(test_images,test_labels) = datasets.cifar10.load_data()
```

# normalize data

```python
train_images,test_images = train_images/255.0 , test_images/255.0
```

```python
class_names = ['aeroplane','car','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i][0]])
    
plt.show()
```

```python
model = Sequential([Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),
                   MaxPooling2D((2,2)),
                   Conv2D(64,(3,3),activation='relu'),
                   MaxPooling2D((2,2)),
                   Conv2D(64,(3,3),activation='relu'),
                   layers.Flatten(),
                   layers.Dense(64,activation='relu'),
                   layers.Dense(10)])

model.summary()
```

```python
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
model.fit(train_images,train_labels,epochs=10,validation_data=(test_images,test_labels),verbose=0)
```

```python
def single_image_predict(model,image_index):
    img =  test_images[image_index]
    true_label = test_labels[image_index][0]
    img_expanded = np.expand_dims(img,axis=0)
    
    prediction = model.predict(img_expanded,verbose=0)
    prediction_label = np.argmax(prediction[0])
    confidence = tf.nn.softmax(prediction[0])
    
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.title(f"prediction:{class_names[prediction_label]} (confidence:{confidence[prediction_label]:.2f})\n" f"true:{class_names[true_label]}")
    
    print(f"prediction : {class_names[prediction_label]}")
    print(f"confidence : {confidence[prediction_label]}")
    
    for i , score in enumerate(confidence):
        print(f"{class_names[i]} : {score*100:.0f}%")
```

```python
image_index = 5
single_image_predict(model,image_index)
```
